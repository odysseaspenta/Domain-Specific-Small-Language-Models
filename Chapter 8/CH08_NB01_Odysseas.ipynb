````python
sequences = protgpt2("<|endoftext|>",
                     max_length=100,
                     do_sample=True,
                     top_k=950,
                     repetition_penalty=1.2,
                     num_return_sequences=10,
                     eos_token_id=0)
for seq in sequences:
  print(seq)
````

## Explanation of the parameters to parameters in the pipeline and the reason for their presence
This code snippet is a standard setup for using ProtGPT2, a generative model specifically trained to "speak" the language of protein sequences. Because protein sequences are just strings of amino acids (A, R, N, D, etc.), they can be modeled using the same Transformer architecture as human languages.Here is the breakdown of each parameter in your protgpt2(...) call:
1. The Input Prompt ("<|endoftext|>" ) This is the "seed" or starting point for the model. In ProtGPT2, <|endoftext|> is the special token used to separate different protein sequences. By starting with this, you are telling the model: "Start a brand new, de novo protein sequence from scratch."
2. max_length=100 The maximum number of tokens (amino acids) the model is allowed to generate. This provides a "hard stop" to prevent the model from generating an infinitely long sequence. Note that this includes the length of your prompt.
3. do_sample=True Enables stochastic (random) sampling. If False, the model always picks the single most likely next amino acid (Greedy Search), which often leads to repetitive or boring sequences. If True, the model picks the next token based on a probability distribution, allowing for more creative and diverse protein designs.
4. top_k=950 A sampling filter that limits the model's choices to the $k$ most likely next tokens. Before picking the next amino acid, the model looks at the top 950 candidates and ignores the rest. Since the "vocabulary" of a protein model is relatively small (20 standard amino acids plus some specials), a top_k of 950 effectively means "consider almost everything" but discard the absolute bottom-tier noise.
5. repetition_penalty=1.2 A discount applied to the probability of tokens that have already appeared. A value $> 1.0$ makes it less likely for the model to repeat the same amino acid or sequence fragment over and over. This is very useful in protein design to avoid unrealistic "stretches" of the same amino acid (like AAAAAAAA...) that might not fold correctly in nature.
6. num_return_sequences=10 The number of independent sequences to generate. Instead of getting just one protein string, the pipeline will run the generation process 10 times and return a list of 10 unique candidate proteins
7. eos_token_id=0 The specific ID for the End Of Sequence token. This tells the model what "stop" looks like. For ProtGPT2, the token with ID 0 corresponds to <|endoftext|>. If the model generates this token before reaching max_length, it will stop immediately, signifying that the protein sequence is naturally complete.

Define a function to calculate the perplexity metric for the generated results.

````python
import torch

def calculate_perplexity(model, tokenizer, text, device):
    encodings = tokenizer(text, return_tensors='pt').to(device)

    input_ids = encodings.input_ids
    target_ids = input_ids.clone()

    with torch.no_grad():
        outputs = model(input_ids, labels=target_ids)

    neg_log_likelihood = outputs.loss

    perplexity = torch.exp(neg_log_likelihood)

    return perplexity
````

In the context of protein generation with models like ProtGPT2, **perplexity** is essentially a measure of how "surprised" the model is by the sequence it is seeing or generating.

Technically, perplexity is the exponentiated average negative log-likelihood of a sequence. If we have a sequence of amino acids $X = (x_1, x_2, ..., x_n)$, the perplexity ($PP$) is calculated as:

$$PP(X) = \exp\left( -\frac{1}{n} \sum_{i=1}^{n} \ln P(x_i | x_{<i}) \right)$$

Here is why a low value for this metric is generally associated with higher quality:

---

### 1. Biological "Grammar" Alignment
Proteins aren't just random strings; they follow strict evolutionary and physical rules (the "grammar" of folding).
* **Low Perplexity:** Means the model assigns a high probability to the sequence. This suggests the generated amino acids follow the patterns, motifs, and conservation rules found in the massive dataset of natural proteins it was trained on.
* **High Perplexity:** Means the model found the sequence "confusing" or unlikely. This often happens when amino acids are placed in an order that would be physically unstable or evolutionarily non-existent.

### 2. Correlation with Foldability
Research in protein language modeling has shown a strong correlation between low perplexity and **structural viability**.
* Proteins with low perplexity are more likely to have a well-defined 3D structure when modeled in tools like AlphaFold.
* High perplexity sequences often result in "disordered" proteins—blobs of amino acids that don't fold into a functional shape and are often degraded by cells.



### 3. Confidence in Local Context
Perplexity measures how well each amino acid fits into the context of the ones before it.
* In a protein, a hydrophobic amino acid (like Leucine) is often followed by other specific residues to form a "core."
* A model with low perplexity "knows" these local chemical environments. If it generates a sequence where every "guess" was highly confident, it’s a sign that the biochemical logic of the sequence is consistent.

---

### The "Perplexity Trap" (A Word of Caution)
While low perplexity is a great indicator of **naturalness**, it is not a perfect metric for **innovation**:

| Perplexity Level | Interpretation | Pros/Cons |
| :--- | :--- | :--- |
| **Extremely Low** | Could be "Memorization." | The model might just be spitting out a sequence it saw in training (e.g., a common sequence like Insulin). |
| **Moderate/Low** | **The "Sweet Spot."** | The sequence looks like a real protein but contains enough variation to potentially be a new, functional design. |
| **High** | Random Noise. | Likely a non-functional sequence that will fail in a lab setting. |

### Summary
Think of perplexity as a **validity filter**. It tells you if the "sentence" the model wrote makes sense in the language of biology. A low perplexity score gives you confidence that the sequence is worth the expensive computational (or wet-lab) resources required to actually build and test it.

Alternatively we can calculate perplexity on a batch of generated protein sequences. Let's define a custom function for this.

````python
protgpt2.tokenizer.pad_token = protgpt2.tokenizer.eos_token

def calculate_batch_perplexity(input_texts, model, tokenizer):
    """
    Calculate perplexity for a batch of input texts using a pretrained language model.

    Args:
    - input_texts (List[str]): A list of input texts to evaluate.

    Returns:
    - List[float]: A list of perplexity scores, one for each input text.
    """
    # Tokenize the batch of texts with padding for uniform length
    inputs = tokenizer(
        input_texts, return_tensors="pt", padding=True, truncation=True
    ).to(model.device)

    input_ids = inputs["input_ids"]
    attention_mask = inputs["attention_mask"]

    # Pass the input batch through the model to get logits
    with torch.no_grad():
        outputs = model(input_ids, attention_mask=attention_mask)
        logits = outputs.logits

    # Shift the logits and input_ids to align targets correctly
    # Logits dimensions are: (batch_size, seq_length, vocab_size)
    shift_logits = logits[:, :-1, :]  # Ignore the last token's logits
    shift_labels = input_ids[:, 1:]   # Skip the first token in the labels

    # Compute log probabilities
    log_probs = torch.nn.functional.log_softmax(shift_logits, dim=-1)

    # Gather the log probabilities for the correct tokens
    target_log_probs = log_probs.gather(dim=-1, index=shift_labels.unsqueeze(-1)).squeeze(-1)

    # Mask out positions corresponding to padding tokens
    target_log_probs = target_log_probs * attention_mask[:, 1:].to(log_probs.dtype)

    # Compute the mean negative log-likelihood for each sequence
    negative_log_likelihood = -target_log_probs.sum(dim=-1) / attention_mask[:, 1:].sum(dim=-1)

    # Compute perplexity for each sequence
    perplexities = torch.exp(negative_log_likelihood)

    # Take mean of perplexities of each batch
    mean_perplexity_score = torch.mean(perplexities)

    return {"perplexities": perplexities, "mean_perplexity": mean_perplexity_score}
````
